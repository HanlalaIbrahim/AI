```python
# importing libraries
import numpy as np
# imprting model module
from sklearn import linear_model

# Pesudo Random Number Generator
rando = np.random.RandomState(0)

# Creating Datas
x = rando.randn(10, 5)
y = rando.randn(10)

# Printing Datas
print('\n x date : \n', x)
print('\n y date : \n', y)

# Creating Model
sdg_reg = linear_model.SGDRegressor(max_iter=1000, penalty="elasticnet", loss='huber', tol=1e-3, average=True)

# Fitting Values to Model
sdg_reg.fit(x, y)

# Printing the Weight Vector
print('\n Weight Vector(s) : ', sdg_reg.coef_)

# Printing Intercept
print('\n Intercept : ', sdg_reg.intercept_)

# Printing Weight Updates
print('\n Number of Weights Updates during Training : ', sdg_reg.t_)
```
