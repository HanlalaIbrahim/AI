```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt

%matplotlib inline
```

```python
data1 = pd.read_csv("5_6264902927333198398.csv")
data1
```

```python
xaxis1 = data1['brand']
xaxis1
```

```python
yaxis1 = data1['price']
yaxis1
```

```python
corrected_price = yaxis1.str.strip('₹').replace(',','',regex=True).astype(int)
corrected_price
```

```python
# Scatter plot
```

```python
pt.scatter(xaxis1,corrected_price)
```

```python
#line plot
```

```python
pt.plot(xaxis1,corrected_price)
```

```python
# bar plot
```

```python
pt.bar(xaxis1,corrected_price)
```

```python
pt.pie(corrected_price,labels=xaxis1)
```
