```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Bubble Chart
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 10, 12]
sizes = [100, 200, 300, 400, 500]  # Bubble sizes

plt.scatter(x, y, s=sizes)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Chart Example')
plt.show()
```

```python
import networkx as nx


# Create a graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, edge_color='black', linewidths=1, font_size=12)
plt.title('Network Diagram Example')
plt.show()
```

```python
import pandas as pd
from pandas.plotting import parallel_coordinates
import matplotlib.pyplot as plt

# Sample data
data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [2, 3, 4, 5],
    'C': [3, 4, 5, 6],
    'D': [4, 5, 6, 7]
})

# Plot parallel coordinates
plt.figure(figsize=(8, 5))
parallel_coordinates(data, 'A')
plt.title('Parallel Coordinates Example')
plt.show()
```

```python
import geopandas as gpd
# Load the data (shapefile)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the choropleth map
world.plot(column='pop_est', cmap='OrRd', legend=True)
plt.title('Choropleth Map Example')
plt.show()
```
