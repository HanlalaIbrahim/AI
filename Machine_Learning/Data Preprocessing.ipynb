```python
data = pd.DataFrame({'Name':['Gladson','Hanlala','Anash','Hasan','Sakthi'],'Salary':[120000,180000,150000,140000,170000],'Pay_lvl':['mid','high','mid','mid','high']})
data
```

```python
data_2= pd.DataFrame({'Salary':[300000,'NaN',2500000,'NaN'],
                    'Age':['NaN',22,'NaN',28]})
data_2
```

```python
imputer = si(missing_values=np.NaN,strategy='mean')
imputer.fit(data_2)
```

```python
mean = imputer.transform(data_2)
mean
```

```python
imputer2 = si(missing_values=np.NaN,strategy='median')
imputer2.fit(data_2)
```

```python
median = imputer.transform(data_2)
median
```
