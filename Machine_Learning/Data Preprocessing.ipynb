```python
data = pd.DataFrame({'Name':['Gladson','Hanlala','Anash','Hasan','Sakthi'],'Salary':[120000,180000,150000,140000,170000],'Pay_lvl':['mid','high','mid','mid','high']})
data
```

```python
data_2= pd.DataFrame({'Salary':[300000,'NaN',2500000,'NaN'],
                    'Age':['NaN',22,'NaN',28]})
data_2
```

```python
imputer = si(missing_values=np.NaN,strategy='mean')
imputer.fit(data_2)
```

```python
mean = imputer.transform(data_2)
mean
```

```python
imputer2 = si(missing_values=np.NaN,strategy='median')
imputer2.fit(data_2)
```

```python
median = imputer.transform(data_2)
median
```

## Normalization

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_3 = pd.DataFrame({'X':np.random.randint(-100,100,200),'Y':np.random.randint(-50,50,200),'Z':np.random.randint(-700,700,200)})
data_3
```

```python
pl = pt.axes(projection = '3d')
```

```python
pl = pt.axes(projection = '3d')
pl.scatter3D(data_3.X,data_3.Y,data_3.Z)
pt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
n = Normalizer()
```

```python
data_4 = n.fit_transform(data_3)
data_4
```

```python
DF = pd.DataFrame(data_4,columns=['X1','Y1','Z1'])
DF
```

```python
pl = pt.axes(projection = '3d')
pl.scatter3D(DF.X1,DF.Y1,DF.Z1)
pt.figure()
```
