```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
sns.lineplot(x=x, y=y)
plt.title('Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Bar plot
x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 25]
plt.figure()
sns.barplot(x=x, y=y)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Histogram
data = np.random.randn(1000)
plt.figure()
sns.histplot(data, bins=30, kde=True)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Box plot
data = np.random.randn(100)
plt.figure()
sns.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```
